```python
import pandas as pd

df = pd.read_csv('poverty.csv')
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = df['PovPct']
y = df['TeenBrth']
plt.scatter(x, y)
plt.xlabel('Percentage of Residents Living Below the Poverty Line')
plt.ylabel('Teen Birth Rate')
plt.title('Teen Birth Rate as a Function of Poverty Level')
plt.show()
```

```python
import numpy as np

n = len(x)
alpha = 0.001
m = np.zeros(n)
b = np.zeros(n)

epoch = 0
max_epochs = 10
mse_values = []

# Iterate to compute m and b
while(epoch < max_epochs):
    predicted = (m * x) + b
    error = predicted - y
    mse = np.sum(error ** 2) / n
    mse_values.append(mse)
    m = m - (alpha * 2 * (np.sum(error * x) / n))
    b = b - (alpha * 2 * (np.sum(error) / n))
    epoch += 1

m = m[0] # Slope
b = b[0] # Y-intercept

# Plot Mean Squared Error
plt.plot(mse_values)
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.show()
```

```python
# Show the equation of the line
print('y = ' + str(m) + 'x + ' + str(b))
```

```python
plt.scatter(x, y)
plt.plot(x, (m * x) + b, color='r')
plt.xlabel('Percentage of Residents Living Below the Poverty Line')
plt.ylabel('Teen Birth Rate')
plt.title('Teen Birth Rate as a Function of Poverty Level')
plt.show()
```
