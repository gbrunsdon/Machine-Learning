# Azure Anomaly Detector

[Azure Anomaly Detector](https://azure.microsoft.com/en-us/services/cognitive-services/anomaly-detector/) is a managed service that detects anomalies in batches or streams of data. Let's it use with a [NASA bearing dataset](https://www.kaggle.com/vinayak123tyagi/bearing-dataset) to predict failures in bearings supporting a high-speed rotating shaft. The dataset includes vibration data for four bearings generated with PCB 353B33 High Sensitivity Quartz ICP accelerometers. The data is time-series data sampled every 10 minutes. Start by loading the dataset and visualizing the data for all four bearings.

```python
import pandas as pd

df = pd.read_csv('Data/bearings.csv', index_col=0, parse_dates=[0])
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()

df.plot(figsize = (12, 6))
```

## Apply Anomaly Detector to one bearing

Bearing #1 failed at the end of the test. Extract the vibration data for that bearing and let Anomaly Detector predict when failure is about to occur.

```python
series = df['Bearing 1']
series.plot(figsize = (12, 6))
```

```python
from azure.ai.anomalydetector import AnomalyDetectorClient
from azure.ai.anomalydetector.models import DetectRequest, TimeSeriesPoint, TimeGranularity
from azure.core.credentials import AzureKeyCredential

KEY = 'SUBSCRIPTION_KEY'
ENDPOINT = 'ANOMALY_DETECTOR_ENDPOINT'

client = AnomalyDetectorClient(AzureKeyCredential(KEY), ENDPOINT)
```

```python
time_series = []

for index, val in series.iteritems():
    time_series.append(TimeSeriesPoint(timestamp=index, value=val))
    
request = DetectRequest(series=time_series, granularity=TimeGranularity.NONE)
response = client.detect_entire_series(request)
```

```python
if any(response.is_anomaly):
    print('An anomaly was detected at index:')
    for i, value in enumerate(response.is_anomaly):
        if value:
            print(i)
else:
    print('No anomalies detected')
```

```python
# Get the index of the first anomaly
i = response.is_anomaly.index(True)
index = pd.DataFrame(series).iloc[i].name

# Plot the time series and show where the first anomaly occurs
series.plot(figsize = (12, 6))
plt.axvline(index, color='r')
```

## Apply Anomaly Detector to all four bearings

Now use principal component analysis (PCA) to combine the data for all four bearings (reducing four dimensions to one) and apply Anomaly Detector to the PCAed dataset. We'll use "normal" data to initialize the PCA transform, and then apply the transform to the remainder of the dataset.

```python
from sklearn.decomposition import PCA

x_train = df['2004-02-12 10:32:39':'2004-02-13 23:42:39']
x_test = df['2004-02-13 23:52:39':]

pca = PCA(n_components=1, svd_solver='full')
x_train_pca = pd.DataFrame(pca.fit_transform(x_train))
x_train_pca.index = x_train.index

x_test_pca = pd.DataFrame(pca.transform(x_test))
x_test_pca.index = x_test.index

series = pd.concat([x_train_pca, x_test_pca]).iloc[:, 0]
series.plot(figsize = (12, 6))
```

```python
client = AnomalyDetectorClient(AzureKeyCredential(KEY), ENDPOINT)

time_series = []

for index, val in series.iteritems():
    time_series.append(TimeSeriesPoint(timestamp=index, value=val))
    
request = DetectRequest(series=time_series, granularity=TimeGranularity.NONE)
response = client.detect_entire_series(request)
```

```python
if any(response.is_anomaly):
    print('An anomaly was detected at index:')
    for i, value in enumerate(response.is_anomaly):
        if value:
            print(i)
else:
    print('No anomalies detected')
```

```python
# This time, plot all anomalies rather than just the first
series.plot(figsize = (12, 6))

for i, value in enumerate(response.is_anomaly):
    if value:
        index = pd.DataFrame(series).iloc[i].name
        plt.axvline(index, color='r', alpha=0.2)
```
